```python
def milli_to_seconds(x):
    if x> (480*1000):
        secs = x/1000 - 480
    if x> (480*1000):
        secs = x/1000 - 480
    if x> (420*1000):
        secs = x/1000 - 420
    if x> (360*1000):
        secs = x/1000 - 360
    if x> (300*1000):
        secs = x/1000 - 300
    if x> (240*1000):
        secs = x/1000 - 240
    elif x> (180*1000):
        secs = x/1000 - 180
    elif x> (120*1000):
        secs = x/1000 - 120        
    elif x>= (60*1000):
        secs =  x/1000 - 60
    elif x < (60*1000):
        secs = x/1000
```
